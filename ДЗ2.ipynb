## **Задача 4**
### 4.1 Сколько в среднем стоят дома, у которых 2 спальни?
### 4.2 Какая в среднем общая площадь домов, у которых стоимость больше 600 000?
### 4.3 Как много домов коснулся ремонт?
### 4.4 Насколько в среднем стоимость домов с оценкой grade домов выше 10 отличается от стоимости домов с оценкой grade меньше 4?


```python
# Сколько в среднем стоят дома, у которых 2 спальни? 
df[df['bedrooms'] ==2]['price'].mean()
```

```python
# Какая в среднем общая площадь домов, у которых стоимость больше 600 000?
df[df['price'] > 600000]['sqft_living'].sum()
```

```python
# Как много домов коснулся ремонт?
df[df['yr_renovated'] > 0]['id'].count()
```

```python
# Насколько в среднем стоимость домов с оценкой grade домов выше 10 отличается от стоимости домов с оценкой grade меньше 4?
c = df[df['grade'] >10]['price'].mean()
n = df[df['grade'] <4]['price'].mean()
m = c-n
print ('Cтоимость домов с оценкой grade домов выше 10 отличается от стоимости домов с оценкой grade меньше 4 отличается на', m) 
```

## **Задача 5**
### 5.1 Выберите дом клиенту Клиент хочет дом с видом на набережную, как минимум с тремя ванными и с подвалом. Сколько вариантов есть у клиента?
### 5.2 Выберите дом клиенту Клиент хочет дом либо с очень красивым видом из окна, либо с видом на набережную, в очень хорошем состоянии и год постройки не меньше 1980 года. В какой ценовом диапазоне будут дома?
### 5.3 Выберите дом клиенту Клиент хочет дом без подвала, с двумя этажами, стоимостью до 150000. Какая оценка по состоянию у таких домов в среднем?


```python
# Клиент хочет дом с видом на набережную, как минимум с тремя ванными и с подвалом. Сколько вариантов есть у клиента?
df[
   (df["waterfront"] > 0) &
   (df["bathrooms"]>= 3) &
   (df["sqft_basement"] > 0 ) 
].count()
```

```python
# Клиент хочет дом либо с очень красивым видом из окна, либо с видом на набережную, в очень хорошем состоянии и год постройки не меньше 1980 года. 
# В какой ценовом диапазоне будут дома?
(df[
    (df['waterfront'] > 0)|
    (df['view'] > 0) &
    (df['grade'] >10) &
    (df['yr_built'] > 1980)
]['price'].min(),
df[
    (df['waterfront'] > 0)|
    (df['view'] > 0) &
    (df['grade'] >10) &
    (df['yr_built'] > 1980)
]['price'].max()     
)
```

```python
# Клиент хочет дом без подвала, с двумя этажами, стоимостью до 150000. Какая оценка по состоянию у таких домов в среднем?
df[
   (df["sqft_basement"] == 0) &
   (df["floors"] == 2) &
   (df["price"] < 150000 ) 
]['grade'].mean() 
```
